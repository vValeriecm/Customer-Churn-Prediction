# 01 - Exploratory Data Analysis (EDA)

**Project**: Customer Churn Prediction & Prevention Dashboard

**Objective**: Understand customer behavior, discover why customers churn, and generate actionable business insights.

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load and Inspect the Data
df = pd.read_csv('Telco-CC.csv')  # Change path if needed
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Data Cleaning
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
df.drop('customerID', axis=1, inplace=True)

# Fix missing TotalCharges
print("Missing TotalCharges before fix:", df['TotalCharges'].isnull().sum())
df['TotalCharges'] = df['TotalCharges'].fillna(0)
print("Missing values after fix:", df['TotalCharges'].isnull().sum())
```

```python
# Churn Rate
print("\nChurn Distribution:")
print(df['Churn'].value_counts())
print("\nChurn Rate (%):")
print(df['Churn'].value_counts(normalize=True) * 100)
```

```python
# Key Plots
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title('Churn by Internet Service')
plt.show()

sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.show()

sns.histplot(data=df, x='tenure', hue='Churn', kde=True, bins=30)
plt.title('Tenure Distribution by Churn')
plt.show()
```

```python
# Additional Plots
cols = ['TechSupport', 'OnlineSecurity', 'PaymentMethod', 'Partner']
for col in cols:
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f'Churn by {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Feature Engineering
df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 100],
                          labels=['New', '1-2 Years', '2-4 Years', '4+ Years'])

services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
            'StreamingTV', 'StreamingMovies']
df['TotalServices'] = df[services].apply(lambda x: (x == 'Yes').sum(), axis=1)

df['HighMonthlyCharge'] = (df['MonthlyCharges'] > 70).astype(int)
df['AvgChargePerMonth'] = df['TotalCharges'] / (df['tenure'] + 1)

print("New features created!")
df[['TenureGroup', 'TotalServices', 'HighMonthlyCharge']].head()
```

## Key Business Insights from EDA

- **Overall Churn Rate**: ~26.5% — This represents a major revenue loss for the company.
- **Contract Type** is the strongest signal: Month-to-month customers churn at ~43%, while 2-year contract customers churn at only ~3%.
- **Fiber optic** users churn significantly more than DSL users.
- Customers with **higher monthly charges** are more likely to churn.
- **New customers** (low tenure) have the highest risk of churning.
- Lack of **Tech Support** and **Online Security** strongly increases churn.
- **Electronic check** users have the highest churn rate.
- Customers **without partners** churn more.
