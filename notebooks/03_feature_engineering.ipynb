# 03 - Feature Engineering

**Project**: Customer Churn Prediction

**Goal**: Create useful new features to improve model performance.

```python
import pandas as pd
import numpy as np

# Load raw data
df = pd.read_csv('Telco-CC.csv')
print("Shape:", df.shape)
```

```python
# Basic Cleaning
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
df.drop('customerID', axis=1, inplace=True)
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
# Feature Engineering

# Tenure Features
df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 100], 
                          labels=['New', '1-2 Years', '2-4 Years', '4+ Years'])

df['IsNewCustomer'] = (df['tenure'] <= 6).astype(int)
df['IsLoyalCustomer'] = (df['tenure'] >= 24).astype(int)

# Charge Features
df['AvgMonthlyCharge'] = df['TotalCharges'] / (df['tenure'] + 1)
df['HighMonthlyCharge'] = (df['MonthlyCharges'] > 70).astype(int)

# Service Features
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
            'StreamingTV', 'StreamingMovies']
df['TotalServices'] = df[services].apply(lambda x: (x == 'Yes').sum(), axis=1)

# Risk Flags
df['MonthToMonth'] = (df['Contract'] == 'Month-to-month').astype(int)
df['ElectronicCheck'] = (df['PaymentMethod'] == 'Electronic check').astype(int)
df['NoTechSupport'] = (df['TechSupport'] == 'No').astype(int)

print("Feature Engineering completed!")
print("Total columns now:", df.shape[1])
```

```python
import os
os.makedirs('../data/processed', exist_ok=True)

# Save enhanced dataset
df.to_csv('../data/processed/df_enhanced.csv', index=False)
```
