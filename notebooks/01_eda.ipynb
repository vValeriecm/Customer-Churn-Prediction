```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df.drop('customerID', axis=1, inplace=True)

print("Dataset Shape:", df.shape)
print("\nChurn Distribution:")
print(df['Churn'].value_counts())
print("\nChurn Rate (%):")
print(df['Churn'].value_counts(normalize=True) * 100)
```

```python
print("\nColumns remaining:", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Check the rows with missing TotalCharges
print("Rows with missing TotalCharges:")
print(df[df['TotalCharges'].isnull()][['tenure', 'MonthlyCharges', 'TotalCharges']])

# Fix: Fill with 0
df['TotalCharges'] = df['TotalCharges'].fillna(0)

print("\nMissing values after fixing:", df['TotalCharges'].isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot 1: Contract Type
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Create new useful features
# Tenure groups
df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 100],
                          labels=['New', '1-2 Years', '2-4 Years', '4+ Years'])

# Total Services subscribed
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
df['TotalServices'] = df[services].apply(lambda x: (x == 'Yes').sum(), axis=1)
```

```python
# High value customer flag
df['HighMonthlyCharge'] = (df['MonthlyCharges'] > 70).astype(int)

# Average monthly charge behavior
df['AvgChargePerMonth'] = df['TotalCharges'] / (df['tenure'] + 1)  # +1 to avoid division by zero

print("New features created!")
print(df[['TenureGroup', 'TotalServices', 'HighMonthlyCharge', 'AvgChargePerMonth']].head())
```
